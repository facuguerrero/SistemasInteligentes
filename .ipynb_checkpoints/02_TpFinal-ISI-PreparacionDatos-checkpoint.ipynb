# Preparaci√≥n de los datos

```python
## Read file
data = pd.read_csv("googleplaystore.csv")
```

## Category

```python
print( len(data['Category'].unique()) , "categories")

print("\n", data['Category'].unique())
```

```python
g = sns.countplot(x="Category",data=data, palette = "Set1")
g.set_xticklabels(g.get_xticklabels(), rotation=90, ha="right")
g 
plt.title('Count of app in each category',size = 20)
```

```python
#Para ver las apps dentro de una categoria
#data[data['Category'] == 'TRAVEL_AND_LOCAL']
```

#### A FAMILY le agrego ENTERTAINMENT, HOUSE_AND_HOME, EDUCATION, TRAVEL_AND_LOCAL

```python
data.Category.replace(to_replace=dict(ENTERTAINMENT="FAMILY"), inplace=True)
data.Category.replace(to_replace=dict(HOUSE_AND_HOME="FAMILY"), inplace=True)
data.Category.replace(to_replace=dict(EDUCATION="FAMILY"), inplace=True)
data.Category.replace(to_replace=dict(TRAVEL_AND_LOCAL="FAMILY"), inplace=True)
```

#### GAME queda igual que antes

#### Recorto la categoria 1.9 que tiene solo una entrada

```python
data = data[data['Category'] != '1.9']
```

#### Reemplazo HEALTH_AND_FITNESS, MEDICAL, SPORTS, BEAUTY

```python
data.Category.replace(to_replace=dict(HEALTH_AND_FITNESS="HEALTH"), inplace=True)
data.Category.replace(to_replace=dict(MEDICAL="HEALTH"), inplace=True)
data.Category.replace(to_replace=dict(SPORTS="HEALTH"), inplace=True)
data.Category.replace(to_replace=dict(BEAUTY="HEALTH"), inplace=True)
```

#### Reemplazo FOOD_AND_DRINK, SOCIAL, COMMUNICATION, EVENTS, DATING

```python
data.Category.replace(to_replace=dict(FOOD_AND_DRINK="SOCIAL"), inplace=True)
data.Category.replace(to_replace=dict(SOCIAL="SOCIAL"), inplace=True)
data.Category.replace(to_replace=dict(COMMUNICATION="SOCIAL"), inplace=True)
data.Category.replace(to_replace=dict(EVENTS="SOCIAL"), inplace=True)
data.Category.replace(to_replace=dict(DATING="SOCIAL"), inplace=True)
```

#### Reemplazo TOOLS, MAPS_AND_NAVIGATION, VIDEO_PLAYERS, PERSONALIZATION

```python
data.Category.replace(to_replace=dict(TOOLS="TOOLS"), inplace=True)
data.Category.replace(to_replace=dict(MAPS_AND_NAVIGATION="TOOLS"), inplace=True)
data.Category.replace(to_replace=dict(VIDEO_PLAYERS="TOOLS"), inplace=True)
data.Category.replace(to_replace=dict(PERSONALIZATION="TOOLS"), inplace=True)
data.Category.replace(to_replace=dict(LIBRARIES_AND_DEMO="TOOLS"), inplace=True)
```

#### Reemplazo el resto

```python
data.Category.replace(to_replace=dict(AUTO_AND_VEHICLES="OTHERS"), inplace=True)
data.Category.replace(to_replace=dict(ART_AND_DESIGN="OTHERS"), inplace=True)
data.Category.replace(to_replace=dict(BOOKS_AND_REFERENCE="OTHERS"), inplace=True)
data.Category.replace(to_replace=dict(BUSINESS="OTHERS"), inplace=True)
data.Category.replace(to_replace=dict(COMICS="OTHERS"), inplace=True)
data.Category.replace(to_replace=dict(FINANCE="OTHERS"), inplace=True)
data.Category.replace(to_replace=dict(LIFESTYLE="OTHERS"), inplace=True)
data.Category.replace(to_replace=dict(SHOPPING="OTHERS"), inplace=True)
data.Category.replace(to_replace=dict(PHOTOGRAPHY="OTHERS"), inplace=True)
data.Category.replace(to_replace=dict(PRODUCTIVITY="OTHERS"), inplace=True)
data.Category.replace(to_replace=dict(PARENTING="OTHERS"), inplace=True)
data.Category.replace(to_replace=dict(WEATHER="OTHERS"), inplace=True)
data.Category.replace(to_replace=dict(NEWS_AND_MAGAZINES="OTHERS"), inplace=True)
```

```python
print( len(data['Category'].unique()) , "categories")

print("\n", data['Category'].unique())
```

```python
g = sns.countplot(x="Category",data=data, palette = "Set1")
g.set_xticklabels(g.get_xticklabels(), rotation=90, ha="right")
g 
plt.title('Count of app in each category',size = 20)
```
